```python
# Step 1: Define Bins & Labels
bins = [0, 100000, 300000, float("inf")]
labels = ["Low", "Medium", "High"]

# Step 2: Create Income Bins
df["Income_Band"] = pd.cut(
    df["Estimated Income"],
    bins=bins,
    labels=labels,
    right=False
)
df["Income_Band"].value_counts() 
# Step 3: Visualize Income Bands
plt.figure(figsize=(8, 5))  
sns.countplot(x="Income_Band", data=df, order=labels)
plt.title("Distribution of Income Bands")   
plt.xlabel("Income Band")
plt.ylabel("Count")
plt.show()  
```

### Bivariate Analysis

```python
import matplotlib.pyplot as plt
import seaborn as sns

predictors = [
    "BRId",
    "GenderId",
    "IAId",
    "Amount of Credit Cards",
    "Nationality",
    "Occupation",
    "Fee Structure",
    "Loyalty Classification",
    "Properties Owned",
    "Risk Weighting",
    "Income_Band"
]

fig, axes = plt.subplots(nrows=4, ncols=3, figsize=(22, 18))
axes = axes.flatten()

for i, predictor in enumerate(predictors):
    sns.countplot(
        data=df,
        x=predictor,
        hue="GenderId",
        ax=axes[i]
    )
    axes[i].set_title(f"{predictor} by Gender")
    axes[i].tick_params(axis='x', rotation=45)

# Remove empty plots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```
